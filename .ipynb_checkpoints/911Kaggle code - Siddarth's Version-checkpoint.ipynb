````python
print len(data['zip'].unique())
print len(data['twp'].unique())
````

````python
data["emergencytype"].unique()
````

````python
data.emergencytype = data.emergencytype.apply(lambda x: x.strip())
````

````python
number = []
for i in range(len(data)):
    if data.emergencytype[i] == "EMS":
        number.append(0)
    elif data.emergencytype[i] == "Fire":
        number.append(1)
    elif data.emergencytype[i] == "Traffic":
        number.append(2)
````

````python
number = pd.DataFrame(number, columns = ["Emergency Type No."])
````

````python
print len(number), len(data)
````

````python
data.set_index("id",inplace = True)
````

````python
data = data.merge(number,how = "outer",left_index = True, right_index = True)
````

````python
data
````

````python
data = data.drop(["desc","e","emergencytype","addr"], axis =1)
data = data.dropna()
````

````python
twpdumdums = pd.get_dummies(data.twp)
twpdumdums
data = data.merge(twpdumdums,how = "outer",left_index=True, right_index = True)
data = data.drop("twp",axis = 1)
````

````python
data
````

````python
X = data[[u'lat', u'lng', u'zip']]
y = data["Emergency Type No."]
````

````python
from sklearn import cross_validation
````

````python
X_train,X_test, y_train,y_test = cross_validation.train_test_split(X,y,test_size = 0.25)
````

````python
from sklearn import linear_model
````

````python
model = linear_model.LogisticRegression()
````

````python
model.fit(X_train,y_train)
````

````python
model.score(X_test,y_test)
````

````python
from sklearn.tree import DecisionTreeClassifier
````

````python
model = DecisionTreeClassifier()
````

````python
model.fit(X_train,y_train)
````

````python
model.score(X_test,y_test)
````
